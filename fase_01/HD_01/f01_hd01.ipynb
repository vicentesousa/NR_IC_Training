# **Hands-on 1: Exemplo cttc-3gpp-channel-simple-ran**

### **Objetivo:**
O exemplo cttc-3gpp-channel-simple-ran.cc permite ao usuário entender como configurar parâmetros básicos de simulações envolvendo a RAN do 5G-LENA.



### **Cenário:**

O cenário construído usa o modelo de canal do 3GPP do TR  38.900 e simula apenas a parte de RAN do NR de uma rede com um UE, um gNb, uma bandwith part e um CC é definido.

## **Requisitos:**

*   Ter instalado o ns-3.36;
*   Ter integrado o 5G-LENA;
*   Ter realizado as leituras preliminares necessárias.



## **Parte 1: Como funciona o Exemplo Lena-Simple-EPC.cc?**


### **Passo 1: Entendendo o código.**

Inicialmente são determinados os includes das bibliotecas a serem utilizadas no exemplo e a descrição do using namespace ns3. 


![1.png](FIGS/1.png)

Em primeiro lugar são estabelecidas as funções g_rxPdcpCallbackCalled e g_rxRlcPDUCallbackCalled que servem como atribuições do callback, entretanto ainda não terão função neste código.


![2.png](FIGS/2.png)

Após isso são estabelecidas algumas funções que serão posteriormente chamadas na main do script. A primeira delas é a função SendPacket, que cria um pacote único e chama a função Send do dispositivo para enviar o pacote até o endereço de destino, se utilizando dos parâmetros: device (dispositivo que enviará o pacote), addr (endereço de destino) e packetSize (tamanho do pacote).

De início ela gera um pacote, após isso inicia o protocolo ipv4 com o Header, e atribui essas propriedades ao pacote (pkt). Em sequência um EpsBearer é adicionado com a tag(1,1) e depois requisitado para entrega através da função Send.


![3.png](FIGS/3.png)

Em seguida é estabelecida uma função para exibir parâmetros do PDU do RLC quando acionada por um callback. Os parâmetros da função são path (caminho do trace source), rnti (identificador temporário de rede de rádio do UE), lcid (ID do canal lógico), bytes (tamanho do PDU do RLC em bytes) e rlcDelay (atraso do PDU do RLC).

![4.png](FIGS/4.png)

Logo após é criada uma função que, quando acionada por um callback, printa no terminal o valor do delay do PDCP. Os parâmetros da função são path (caminho do trace source), rnti (identificador temporário de rede de rádio do UE), lcid (ID do canal lógico), bytes (tamanho do PDU do RLC em bytes) e pdcpDelay (atraso do PDCP).

![5.png](FIGS/5.png)

Por último, é estabelecida uma função que conecta os traces do PDCP e do RLC com os seus respectivos trace sources. 

![6.png](FIGS/6.png)

Com isso, a main é iniciada e são definidas as variáveis que serão utilizadas em outras partes do script.

![7.png](FIGS/7.png)

As linhas de código abaixo estabelecem as variáveis que poderão ser alteradas via terminal.

![8.png](FIGS/8.png)

A seguir, utiliza-se o GridScenarioHelper para posicionar as antenas. Inicialmente, define-se o grid das gNbs cujas colunas variam de acordo com o número de antenas, neste caso, uma. Após isso, determina-se a distância horizontal entre as antenas, caso exista mais de uma. Assim, configura-se a altura da antena das gNbs e dos UEs.

Na linha gridScenario.SetSectorization (GridScenarioHelper::SINGLE) está sendo definido o “tipo” de antena, se ela será setorizada ou omnidirecional. Além de SINGLE (cujo setor possui 360º), também pode-se definir como TRIPLE (3 setores de 120°) ou NONE (valor não configurado).

Após isso é determinado o número de gNbs e UEs e posteriormente cria-se cria-se um grid 3x3 onde os UEs serão distribuídos. A linha randomStream += gridScenario.AssignStreams (randomStream) atribui um número de fluxo de variável aleatória fixo às variáveis ​​aleatórias usadas por este modelo. Retorna o número de fluxos (possivelmente zero) que foram atribuídos. Por fim, o cenário é criado de fato.


![9.png](FIGS/9.png)

Assim, instância-se o NrPointToPointEpcHelper, o IdealBeamformingHelper e o NrHelper, para construir a estrutura básica da rede NR.

![10.png](FIGS/10.png)

Com isso, cria-se as bandwith parts da rede, no caso, apenas uma BWP. Logo em seguida configura-se a BWP usando a mínima configuração requerida (SimpleOperationBandConf), passando a frequência central, a largura da banda, o número de CCs da banda operacional e o modelo de propagação do 3GPP como parâmetros do método.

![11.png](FIGS/11.png)

Por conseguinte, é configurado o período de update em 0 milisegundos, a atribuindo os valores de Mcs do Downlink Fixo como True, Mcs de Downlink Inicial como 28, período de update em 0 e Shadowing como False. Com isso é inicializada a BWP da banda 1, selecionado o beamforming do tipo DirectPath; selecionado o número de colunas, linhas e tipo das antenas das UEs e gNbs com o método SetUe/GnbAntennaAttribute; são instalados os devices nos nós com o NetDeviceContainer; e assinalados como random as streams para cada device, com o AssignStreams.

![12.png](FIGS/12.png)

Após zerar os atributos de enbNetDev e a numerologia da BWP1, o codigo se utiliza de uma função for para atualizar essa configurações de cada gNb e UE do sistema, depois cria a internet e gera os IPV4s com o epcHelper.

Além disso, também se dá attach entre os UEs e gNbs mais próximos entre si.
